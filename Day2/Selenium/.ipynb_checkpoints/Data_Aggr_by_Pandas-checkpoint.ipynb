```python
import json
data = json.load(open('titanic-passengers.json'))
jsonDF = pd.DataFrame([])
jsonDF.info()
```

```python
for d in data:
    #print(d['fields'])
    #print(pd.json_normalize(d['fields']))
    jsonDF = jsonDF.append([pd.json_normalize(d['fields'])])
jsonDF.info()
```

# So let's remove some columns, then try to test :)

```python
del csvDF['Age']
del csvDF['Ticket']
csvDF.info()
```

```python
csvDF.head()
```

```python
del jsonDF['fare']
del jsonDF['cabin']
jsonDF = jsonDF.rename(columns={'name': 'Name', 'embarked': 'Embarked', 'age': 'Age',
                                'parch': 'Parch', 'pclass': 'Pclass', 'sex': 'Sex',
                                'survived': 'Survived', 'ticket': 'Ticket', 
                                'passengerid': 'PassengerId', 'sibsp': 'SibSp'})
jsonDF.head()
```

```python
csvDF.to_csv('dataset1.csv', index=False)
jsonDF.to_csv('dataset2.csv', index=False)
```

```python
mergedDF = pd.merge(csvDF, jsonDF, 
                    on=["PassengerId","Survived","Pclass","Name","Sex","SibSp","Parch","Embarked"])
mergedDF.info()
```
