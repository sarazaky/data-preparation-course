```python
import pandas as pd
```

```python
df1 = pd.read_csv('epldata_final.csv')
df1.info()
```

```python
df2 = pd.read_csv('Premier_League_Player_Top_Scorers.csv')
df2.info()
```

```python
df1.head()
```

```python
df2.head()
```

```python
innerMergedDF = pd.merge(df1, df2, on=["Player"])
innerMergedDF.info()
```

```python
innerMergedDF.head(20)
```

```python
outerMergedDF = pd.merge(df1, df2, on=["Player"], how='outer')
outerMergedDF.info()
```

```python
outerMergedDF.head(50)
```

```python
innerMergedDF.to_csv("inner_merged_output.csv", index=False)
outerMergedDF.to_csv("outer_merged_output.csv", index=False)
```
