```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd

driver = webdriver.Firefox()
driver.get('https://www.premierleague.com/stats/top/players/goals?se=79')
for i in range(100000000):
        n = 2+3
players = driver.find_elements_by_xpath('//a[@class="playerName"]')
players_list = []
for p in players:
    players_list.append(p.text)
print(players_list)  

nationality = driver.find_elements_by_xpath('//span[@class="playerCountry"]')
nationality_list = []
for n in nationality:
    nationality_list.append(n.text)
print(nationality_list) 


data_tuples = list(zip(players_list,nationality_list))
df1 = pd.DataFrame(data_tuples, columns=['name','nationality'])   

driver.close()

df1.to_csv("players_data.csv", index=False)
```
